```python
# Importing libraries
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

# Functions Needed

```python
def initalization(inp,out):
    inp = QuantumRegister(inp,'q')
    out = ClassicalRegister(out,'c')
    cir = QuantumCircuit(inp,out)
    return inp, out, cir
```

```python
def simulation(cir):
    aer = AerSimulator()
    res = aer.run(cir,shots = 1024).result()
    cnt = res.get_counts()
    return [cnt]
```

# Question 1

![My Image](questions/q1.jpeg)

```python
inp1, out1, cir1 = initalization(2,4)

cir1.id(inp1[0])
cir1.x(inp1[1])

for i in range(2):
    cir1.h(inp1[i])

cir1.barrier()

cir1.cx(inp1[0],inp1[1])

cir1.barrier()
cir1.h(inp1[0])
cir1.measure(inp1[0],out1[0])
cir1.draw('mpl')
```

```python
cnt1 = simulation(cir1)
plot_histogram(cnt1)
```
